```python
# Importando bibliotecas
import pandas as pd
# Mágica do matplotlib
%matplotlib inline
```

```python
# Carregando os dados do arquivo para o DataFrame
df = pd.read_csv('cursos-prouni.csv')

# Exibe as primeiras linhas
df.head(3)
```

```python
# Verificando o tamanho do DataFrame, 
# Formato (linhas, colunas)
df.shape
```

```python
# Os tipos de dados das colunas
df.dtypes
```

```python
# Analise matematica das colunas numericas
df.describe()
```

```python
# Descrever os dados
df.info()
```

```python
# Quantidade de cursos por UF
df['uf_busca'].value_counts()
```

```python
# Qual o curso com a mensalidade mais alta
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
# Qual e o curso com a menor mensalidade
df[df['mensalidade'] == df['mensalidade'].min()]
```

```python
# Qual a mensalidade mais cara por UF?
df.groupby('uf_busca')['mensalidade'].max()
```

```python
# Qual a mensalidade mais barata por UF?
df.groupby('uf_busca')['mensalidade'].min()
```

```python
# Mostrar todos os dados dos cursos mais caros de cada UF
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()]
```

```python
# Todos os cursos onde a mensalidade é entre 500 e 1000 reais
df[(df['mensalidade'] > 500) & (df['mensalidade'] < 1000)]
```

```python
# Quais as 5 maiores mensalidades
df.nlargest(5, 'mensalidade')
```

```python
# Quais as 5 menores mensalidades
df.nsmallest(5, 'mensalidade')
```
